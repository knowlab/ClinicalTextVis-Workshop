# Hackathon: data visualization for clinical text

Welcome to this session! Let's visualise and celebrate!

The reason for this hackathon is to share ideas and expertise for data visualisation, expecially for natural language process of clinical text. Data visualization is something we have to do, but not necessarily something we reflect on often. It is also challenging both in terms of skill and art. Therefore, we hope this would be a good opportunity for us to collabrate and tackle.

## Basic ideas

The aim of this hackathon is to create one piece (or more if you like) of visualization that could help us understand, screen or analyse clinical texts (or in general any text). 

Feel free to try anything, bring in your own model or data!

When we talk about visualization, we could mean visualization of different things including but not limited to:

* Visualisation of data(text) itself: e.g. length of documents/word frequencies
* Visualisation of models: e.g. parameters in the attention mechanism
* visualisation of outputs or downstream analysis

We have prepared some data and model outputs to give us a head start. In the following blocks, we will guide you through it.
