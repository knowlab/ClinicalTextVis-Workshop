### Example 3: Model weight visualisation
Modelling the multi-head self-attention weights in Transformer for a sentence in the report.

```python
!pip install bertviz 
# acknowledgement to the nice package from jessevig! https://github.com/jessevig/bertviz
from bertviz import model_view
```

```python
#Load model and tokeniser
from transformers import AutoTokenizer, AutoModel

#specify your preferred model here! 
#model_version = 'bert-base-uncased'
model_version = 'bionlp/bluebert_pubmed_mimic_uncased_L-12_H-768_A-12'
#model_version = "microsoft/BiomedNLP-PubMedBERT-base-uncased-abstract-fulltext"

do_lower_case = True
model = AutoModel.from_pretrained(model_version, output_attentions=True)
tokenizer = AutoTokenizer.from_pretrained(model_version,do_lower_case=do_lower_case)
#print(tokenizer.get_vocab())
```

```python
# Retrieve sentence from the dataset
# randomly define a doc by the ids here
subj_id = '0'
row_id = '0'
sent_ind = 6

import sys
github_path = './ClinicalTextVis-Workshop'
sys.path.append(github_path)
from utils import read_JSON_file, get_sentences_offset_from_SemEHR
import pandas as pd

# retrieve the sentence
json_doc_file_name = 'doc-%s-%s.json' % (subj_id, row_id)
json_str = read_JSON_file('%s/SemEHR_processed_jsons/%s' % (github_path, json_doc_file_name))
list_sent_offsets = get_sentences_offset_from_SemEHR(json_str)
sent_ind = sent_ind if sent_ind + 1 < len(list_sent_offsets) else -2 # make sure the sent_ind is within the number of sentences identified
sent_offset = list_sent_offsets[sent_ind]
sent_next_offset = list_sent_offsets[sent_ind+1]

#loop over the document csv file
df = pd.read_csv('%s/noteevents.csv' % github_path)
doc_specified = df['text'][int(row_id)]              
sent = doc_specified[sent_offset[0]:sent_offset[1]]
sent_next = doc_specified[sent_next_offset[0]:sent_next_offset[1]]
print(sent, sent_next)
```

```python
# Load attention weights
# tokenise the two sentences together, connected with a special token [SEP]
inputs = tokenizer.encode_plus(sent, sent_next, return_tensors='pt', add_special_tokens=True)
token_type_ids = inputs['token_type_ids']
input_ids = inputs['input_ids']
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
```
